#### 1.2.Exploratory Data Analysis

##### Check for Distributions

```python
numerical_cols = data.select_dtypes(include=['float64']).columns
```

```python
# Calculate number of rows and columns for subplots
num_cols = len(numerical_cols)
num_rows = (num_cols + 3) // 4  # Calculate number of rows needed (4 plots per row)

# Create subplots
fig, axes = plt.subplots(num_rows, 4, figsize=(25, num_rows*4))  # Adjust figsize as needed

# Flatten axes for easy iteration
axes = axes.flatten()

# Plot each numerical variable
for i, col in enumerate(numerical_cols):
    sns.histplot(data[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

for j in range(num_cols, num_rows * 4):
    axes[j].axis('off')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
skewness = data[numerical_cols].skew()
print("Skewness:")
print(skewness)
```

```python
kurtosis = data[numerical_cols].kurtosis()
print("\nKurtosis:")
print(kurtosis)
```

```python
# Assuming 'diagnosis' is the categorical variable of interest
categorical_col = 'diagnosis'

# Plot the distribution of the categorical variable using hue
plt.figure(figsize=(3, 2))
sns.countplot(x=categorical_col, data=data, hue=categorical_col, palette='viridis', legend=False)
plt.title(f'Distribution of {categorical_col}')
plt.xlabel(categorical_col)
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

##### Test for Skewness

```python
from scipy.stats import shapiro
```

```python
# Perform Shapiro-Wilk test
shapiro_test = shapiro(data['radius_se'])

print(f"Shapiro-Wilk Test - p-value: {shapiro_test.pvalue}")
```

```python
# Log transformation of the variable
data['radius_se_lt'] = np.log(data['radius_se'])

# Check the distribution of the transformed variable
sns.histplot(data['radius_se_lt'], kde=True)
plt.title("Distribution of Transformed Variable (Log Transformation)")
plt.xlabel("Transformed Variable")
plt.ylabel("Frequency")
plt.show()
```

```python
# Replace the original variable 'variable' with the transformed variable 'transformed_variable'
data['radius_se'] = data['radius_se_lt']

# Optionally, you can drop the 'transformed_variable' column if no longer needed
data.drop('radius_se_lt', axis=1, inplace=True)
```

### 2. Feature Engineering

##### Feature Selection

##### X,Y Split

```python
Y = data.iloc[:, 0]  # Target variable
X = data.iloc[:, 1:]  # Predictor variables 
```

```python
X
```

```python
Y
```

##### Export X Stats

```python
description = X.describe()
```

```python
variable_names = description.columns.tolist()
max_values = description.loc['max'].tolist()
min_values = description.loc['min'].tolist()
```

```python
summary_df = pd.DataFrame({
    'Variable Name': variable_names,
    'Max': max_values,
    'Min': min_values
})
```

```python
import pickle
```

```python
pickle.dump(summary_df,open('summary.pkl','wb'))
```

##### Feature Scaling (Normalization)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
import pickle
pickle.dump(scaler,open('scaler.pkl','wb'))
```

##### Train and Testing Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,Y,test_size=0.3, random_state=42)
```

```python
X_train
```

### 3. Model Building

##### Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

#### Model Evaluation

##### Regression Coefficients

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
# Define colors based on the relationship between y_test and reg_pred
colors = np.where(np.abs(y_test - y_pred) < 0.1, 'green', 'red')  # Green if close, red otherwise

# Plot test data with predictions
plt.figure(figsize=(4, 3))
plt.scatter(y_test, y_pred, c=colors)

# Customize plot labels and title
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (reg_pred)')
plt.title('Actual vs. Predicted Values Scatter Plot')

# Display plot
plt.show()
```

```python
# Plot predicted vs actual values
plt.figure(figsize=(4, 3))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)  # Scatter plot of actual vs predicted values

# Plotting the perfect predictions line (y = x)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  

# Customize plot labels and title
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (reg_pred)')
plt.title('Actual vs. Predicted Values Scatter Plot')

# Show plot
plt.show()
```

###### Note:

In this scatter plot, The blue dots represent individual data points where the x-coordinate is the actual value (y_test) and the y-coordinate is the predicted value (reg_pred).

The red dashed line (y = x) represents perfect predictions, where the predicted value equals the actual value.

###### Interpretation:
Data points close to the red dashed line (y = x) indicate accurate predictions, where the predicted values closely match the actual values.
Points that deviate far from the red dashed line represent prediction errors, where the model's predictions differ significantly from the actual values.

```python
from sklearn.metrics import accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```
